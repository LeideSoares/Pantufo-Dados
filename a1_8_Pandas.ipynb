## Introdução ao Pandas<a id="Intro"></a>
Neste tutorial, vamos explorar o Pandas, uma poderosa biblioteca de análise de dados em Python. O Pandas é amplamente utilizado para manipulação e análise de dados, permitindo que você realize uma variedade de tarefas relacionadas a dados. Vamos cobrir os seguintes tópicos:

## O que é o Pandas?
O Pandas é uma biblioteca Python de código aberto que fornece estruturas de dados e funções para trabalhar com dados de forma eficiente. As principais estruturas de dados do Pandas são Séries, DataFrames e Painéis. Ele é amplamente utilizado em análise de dados e é uma ferramenta essencial para cientistas de dados e analistas.

## Estruturas de Dados no Pandas<a id="Estrutura-dados"></a>

### Séries
Uma Série é uma estrutura unidimensional que pode conter dados de diferentes tipos. Você pode criar uma Série da seguinte maneira:

```python
dados = [1, 2, 3, 4, 5]

serie = pd.Series(dados)
```

```python
serie
```

```python
type(serie)
```

### DataFrames
Um DataFrame é uma estrutura bidimensional semelhante a uma planilha. Ele é a estrutura de dados mais comumente usada no Pandas. Você pode criar um DataFrame assim:

```python
dados = {'Nome': ['Alice', 'Bob', 'Charlie', 'David'],
        'Idade': [25, 30, 35, 40]}

df = pd.DataFrame(dados)
```

```python
df
```

### Painéis
Painéis são estruturas tridimensionais que raramente são usadas em análise de dados. Normalmente, Séries e DataFrames são suficientes para a maioria das tarefas.

## Limpeza e Preparação de Dados<a id="limpeza"></a>

### Tratamento de valores ausentes
O Pandas fornece métodos para lidar com valores ausentes, como dropna() para remover linhas ou colunas com valores ausentes e fillna() para preencher valores ausentes.

```python
df.isna().sum()
```

```python
df.dropna()  # Remove linhas com valores ausentes
```

```python
df.fillna(0)  # Preenche valores ausentes com 0
```

### Remoção de duplicados
Para remover duplicados de um DataFrame, use `drop_duplicates()`.

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

### Transformação de dados
Você pode aplicar funções a colunas de dados para transformá-los.

```python
df['Age'] = df['Age']  + 10 # Multiplica os valores da coluna 'Idade' por 2
```

```python
df['Age']
```

### Normalização e padronização
A normalização e a padronização são técnicas comuns para preparar dados. Use funções como `min-max scaling` ou `z-score normalization`.

```python
df['Idade_2023'] =  df['Age'] + 10
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = MinMaxScaler()
df['Idade_2023_mms'] = scaler.fit_transform(df[['Idade_2023']])
```

```python
df['Idade_2023'].plot(kind='hist', bins=20, edgecolor='white')
```

```python
df['Idade_2023_mms'].plot(kind='hist', bins=20, edgecolor='white')
```

## Agrupamento e Pivotagem de Dados<a id="agrupamento"></a>

### Agrupando dados
Agrupe dados com base em critérios específicos usando `groupby()`.

```python
df.groupby(['Sex', 'Embarked'])[['Age']].mean()
```

### Pivotagem de tabelas
A pivotagem permite reorganizar um DataFrame de uma forma mais adequada para análise.

```python
df.pivot_table(index='Sex', columns='Embarked', values='Age')
```

## Mesclagem e Junção de Dados<a id="juncao"></a>

### Concatenação de DataFrames
Você pode combinar DataFrames usando `concat()`.

```python
# Para teste - APENAS

df_copia = titanic.loc[300:400]
```

```python
pd.concat([df, df_copia], axis=0)
```

```python
pd.concat([df, df_copia], axis=1)
```

```python
df_append = df.append(df_copia)
```

```python
df_append
```

### Mesclagem por chave
Realize mesclagens de DataFrames com base em chaves usando `merge()`.

```python
pd.merge(df, df_copia, on='Pclass')
```

### Junção de tabelas
Junte tabelas com base em índices ou colunas usando `join()`.

```python
df.join(df_copia, on='Pclass', lsuffix='_dados_esquerda', rsuffix='_nacionalidade')
```
