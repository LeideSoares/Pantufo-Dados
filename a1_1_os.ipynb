## 4. Caminhos de Arquivo <a id="de-Arquivo"></a>
[Para cima](#indice)

### 4.1 Construção de caminhos de arquivo
A função _**os.path.join()**_ é usada para construir caminhos de arquivo de maneira segura e independente do sistema operacional:

```python
caminho = os.path.join("/paises_lusofono", "brasil", "relatorio.txt")
```

```python
print(caminho)
```

### 4.2 Separação de caminhos
Você pode separar caminhos de arquivo em diretório e nome de arquivo com **`os.path.split()`**:

```python
diretorio, nome_do_ficheiro = os.path.split("/paises_lusofono/brasil/relatorio.txt")
```

```python
print("Diretório:", diretorio)
print("Nome do arquivo:", nome_do_ficheiro)
```

### 4.3 Resolução de caminhos
Para resolver caminhos relativos em caminhos absolutos, use **`os.path.abspath()`**:

```python
caminho_absoluto = os.path.abspath("paises_lusofonos/Africa")
```

```python
print(caminho_absoluto)
```

## 4.5 Exploração Recursiva de Diretórios
**`os.walk`** $-->$ é uma função em Python que gera os nomes dos arquivos em um determinado diretório e em todos os seus subdiretórios, de forma recursiva. Ela retorna um gerador que produz tuplas contendo três elementos para cada diretório percorrido:

- O caminho para o diretório atual.
- Uma lista dos subdiretórios encontrados no diretório atual.
- Uma lista dos arquivos encontrados no diretório atual.

```python
for caminho, diretorios, ficheiros in os.walk('C:/PantufoData/a3DataScience/a_Pantufo_Dados'):
    print('Caminho atual:  ', caminho)
    print('Diretorios:  ', diretorios)
    print('Ficheiros:  ', ficheiros)
    print()
```

### 5.1 Leitura e escrita de arquivos
Você pode abrir arquivos para leitura ou escrita com funções incorporadas do Python, como **`open()`**, e usar os métodos   **`write()`** e **`read()`**:


- Texto da [Agencia Brasil](https://agenciabrasil.ebc.com.br/internacional/noticia/2022-05/lingua-portuguesa-e-quarta-mais-falada-no-mundo)

```python
texto = """260 milhões de pessoas (3,7% da população mundial) falam o português, o quarto mais usado, depois do mandarim, inglês e espanhol"""
```

```python
with open("relatorio_lusofonia.txt", "w") as file:
    file.write(texto)
    
```

```python
with open("relatorio_lusofonia.txt", "r") as file:
    conteudo = file.read()
    
```

```python
print(conteudo)
```

### 5.2 Verificação de existência de arquivos
Você pode verificar se um arquivo existe com **os.path.exists()**:

```python
if os.path.exists("relatorio_lusofonia.txt"):
    print("O arquivo existe.")
else:
    print("O arquivo não existe.")
```

```python
import geopandas as gpd
import os
import matplotlib.pyplot as plt

# Ler os Dados Geograficos
geo_dados = gpd.read_file('../data/data_africa_boundaries/Africa_Boundaries.shp')



nome_diretorio_paises = 'PAISES_AFRICANOS_LUSOFONOS'

# Criar a Pasta (diretorio)
os.mkdir(nome_diretorio_paises)
os.chdir(os.path.abspath(nome_diretorio_paises))

# Nomes dos paises
paises = [('Angola', 'AGO'),  ('Cabo Verde', 'CPV'), ('Guiné-Bissau', 'GNB'), 
          ('Moçambique', 'MOZ'),  ('São Tomé e Príncipe', 'STP')]
```

```python
# Ler iterativamente os paises
for pais in paises:
    
    # Criar diretororio com Nome do Pais iterado
    os.mkdir(pais[0])
    
    
    
    geo_dados[geo_dados['ISO'] == pais[1]].plot(color='#3333cc');
    plt.title(pais[0])
    plt.savefig(pais[0] + '/mapa_' + pais[0] )
    plt.show();
```
