## 1. Introdução ao SQLite3 <a id="ao-SQLite3"></a>
[Para cima](#indice)

### 1.1 O que é o SQLite3?
O SQLite3 é um sistema de gerenciamento de banco de dados relacional que é autocontido, sem servidor e de código aberto. Ele é amplamente utilizado devido à sua simplicidade, eficiência e capacidade de gerenciar bancos de dados embutidos. O SQLite3 é uma biblioteca C que pode ser acessada por várias linguagens de programação, incluindo Python.

### 1.2 Vantagens e Recursos
Sem necessidade de servidor: Não requer um servidor dedicado; os dados são armazenados em um único arquivo.
Portátil: Pode ser executado em várias plataformas.
Transações ACID: Suporta transações ACID (Atomicidade, Consistência, Isolamento e Durabilidade).
Rápido e eficiente: Ideal para aplicações que requerem acesso rápido a pequenos bancos de dados.
Embutido: Pode ser integrado diretamente em aplicativos.

### 3.2 Consulta de Dados
Para consultar dados, utilize a instrução SQL SELECT:

```python
# Define um comando SQL para consultar todos os registros
selecionar_tudo_sql = '''SELECT * FROM tarefas'''

# Executa o comando SQL e recupera os resultados
cursor.execute(selecionar_tudo_sql)
resultados = cursor.fetchall()

for resultado in resultados:
    print(resultado)
```

## 4. Consultas SQL <a id="Consultas-SQL"></a>
[Para cima](#indice)

### 4.1 Seleção de Dados
Você pode especificar colunas específicas em uma instrução SELECT para recuperar apenas os dados necessários.

```python
# Define um comando SQL para selecionar colunas específicas
selecionar_colunas_sql = '''SELECT pais FROM tarefas'''

# Executa o comando SQL e recupera os resultados
cursor.execute(selecionar_colunas_sql)
resultados = cursor.fetchall()

for resultado in resultados:
    print(resultado[0])
```

### 4.2 Filtros e Condições
Use a cláusula WHERE para aplicar filtros a consultas:

```python
# Define um comando SQL com filtro
filtrar_dados_sql = '''SELECT * FROM tarefas WHERE pais = ?'''
filtro = ('Brasil',)

# Executa o comando SQL com o filtro
cursor.execute(filtrar_dados_sql, filtro)
resultados = cursor.fetchall()

for resultado in resultados:
    print(resultado)
```

### 4.3 Ordenação de Resultados
Você pode ordenar os resultados de uma consulta usando a cláusula ORDER BY:

```python
# Define um comando SQL com ordenação
selecionar_ordenado_sql = '''SELECT * FROM tarefas ORDER BY pais ASC'''

# Executa o comando SQL com a ordenação
cursor.execute(selecionar_ordenado_sql)
resultados = cursor.fetchall()

for resultado in resultados:
    print(resultado)
```

### 5.1 Exclusão de Registros
Para excluir registros, utilize a instrução SQL DELETE:

```python
# Define um comando SQL para excluir registros
excluir_dados_sql = '''DELETE FROM tarefas WHERE id = ?'''
registro_a_excluir = (1,)

# Executa o comando SQL
cursor.execute(excluir_dados_sql, registro_a_excluir)

# Confirma as alterações no banco de dados
conn.commit()
```

## 6. Consultas Avançadas <a id="Consultas-Avançadas"></a>

### 6.1 Junção de Tabelas
O SQLite3 suporta junção de tabelas para combinar dados de várias tabelas em uma única consulta:

```python
# Define um comando SQL com junção
select_join_sql = '''SELECT tarefas.descricao, categorias.nome
                    FROM tarefas
                    JOIN categorias ON tarefas.categoria_id = categorias.id'''

# Executa o comando SQL
cursor.execute(select_join_sql)
resultados = cursor.fetchall()

for resultado in resultados:
    print(resultado)
```

### 6.2 Funções de Agregação

## 7. Uso de Parâmetros <a id="de-Parâmetros"></a>

### 7.1 Consultas Parametrizadas
Para evitar a injeção de SQL e tornar seu código mais seguro, use consultas parametrizadas:

```python
# Define um comando SQL parametrizado
param_sql = '''SELECT * FROM tarefas WHERE concluida = ? AND descricao LIKE ?'''

# Executa o comando SQL com parâmetros
params = (True, '%comprar%')
cursor.execute(param_sql, params)
resultados = cursor.fetchall()
```

### 7.2 Prevenção contra Injeção SQL
Ao usar consultas parametrizadas, você evita a injeção de SQL, tornando seu código mais seguro.

### 7.3 Uso de Placeholders
Placeholders, como ? e :nome, são usados para indicar onde os valores dos parâmetros devem ser inseridos em uma consulta.

```python
dados_dos_cantores = []

URL_BASE = 'https://www.last.fm'

for numeracao_pagina in range(1,26):
    URL_PAGINACAO = f'https://www.last.fm/pt/tag/kizomba/artists?page={numeracao_pagina}'
    
    soupa = BS(requests.get(URL_PAGINACAO).content, 'html.parser')
    
    soupa = soupa.find_all(class_ = 'link-block-target')
    
    for cantor in soupa:
        
        dados_dos_cantores.append([cantor.text, URL_BASE + cantor['href']])
        
        
        
dados_dos_cantores
```

# CRIAR e SALVAR na BASE DADOS

```python
conetar = sqlite3.connect('CANTORES_KIZOMBA.db')
```

```python
codigo_criar_tabela =  """
    CREATE TABLE IF NOT EXISTS cantores (
     id INTEGER PRIMARY KEY,
     nome TEXT(80),
     url TEXT(255)
     )
"""

cursor = conetar.cursor()

cursor.execute(codigo_criar_tabela)

conetar.commit()
```

```python
# INSERIR INFORMACAO

codigo_inserir_dados = """
    INSERT INTO cantores (nome, url) VALUES (?, ?)
"""

for dados_cantor in dados_dos_cantores:
    
    dados = (dados_cantor[0], dados_cantor[1])
    
    cursor.execute(codigo_inserir_dados, dados)
    
conetar.commit()
```
